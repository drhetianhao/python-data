```python
# to install wordcloud - execute "pip install wordcloud" from Anaconda prompt

# read the course descriptions
with open("Course-Descriptions.txt", "r") as fh:
    filedata = fh.read()

# check contents
print("File data sample : ", filedata[:200])

# Create stopword list:
from wordcloud import WordCloud, STOPWORDS

stopwords = set(STOPWORDS)

# Generate wordcloud data
wordcloud = WordCloud(
    stopwords=stopwords, max_words=25, background_color="white"
).generate(filedata)
```

## 01_03 Displaying the Wordcloud

```python
import matplotlib.pyplot as mpLib

mpLib.imshow(wordcloud)
mpLib.axis("off")
mpLib.show()
```

## 01_04 Enhancing Wordcloud

```python
# Add more words to ignore
stopwords.update(["many", "using", "want", "value"])

# Redo stop words. Limit number of words
wordcloud = WordCloud(
    stopwords=stopwords, max_words=10, background_color="azure"
).generate(filedata)

mpLib.imshow(wordcloud)
mpLib.axis("off")
mpLib.show()
```
